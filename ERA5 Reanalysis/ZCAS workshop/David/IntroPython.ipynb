```python
import cartopy.crs as ccrs

f=plt.figure(figsize=(10,5))

# Projections
# https://scitools.org.uk/cartopy/docs/latest/crs/projections.html

# Colormaps
# https://matplotlib.org/tutorials/colors/colormaps.html

proj=ccrs.PlateCarree()
ax1 = plt.subplot(121, projection=proj)
ax1.coastlines(resolution='110m')
ax1.set_extent([-90, -30, -55, 15], ccrs.PlateCarree())
p = ax1.contourf(lon, lat, v[0,:,:],
               extend='both',
               cmap = plt.get_cmap('inferno'),
               transform=ccrs.PlateCarree())
cbar = plt.colorbar(p, orientation='vertical', shrink=0.8)
cbar.set_label('m/s')
ax1.set_title('PlateCarree')

proj=ccrs.Orthographic(central_longitude=-60, central_latitude=-20, globe=None)
ax2 = plt.subplot(122, projection=proj)
ax2.coastlines(resolution='110m', linewidth=0.5)
plt.contourf(lon, lat, v[1,:,:],
               extend='both',
               cmap = 'seismic',
               transform=ccrs.PlateCarree())
plt.colorbar(orientation='vertical', shrink=0.8)
cbar.set_label('m/s')
ax2.set_title('Orthographic')
ax2.set_global()

plt.tight_layout()
plt.show()
```

## Médias dos dias do Evento de ZCAS

27 a 30 de Outubro de 2009

```python
myv = v[np.where(time.year==2009)[0],:,:]
myu = u[np.where(time.year==2009)[0],:,:]
```

```python
myv_media = np.mean(myv, axis=0)
```

```python
myv_media.shape
```

```python
f=plt.figure(figsize=(10,6))

proj=ccrs.Orthographic(central_longitude=-60, central_latitude=-20, globe=None)
proj=ccrs.PlateCarree()
ax = plt.subplot(111, projection=ccrs.PlateCarree())
ax.coastlines(resolution='110m', linewidth=0.5)
plt.contourf(lon, lat, myv_media,
               extend='both',
               cmap = 'RdBu_r',
                levels=np.arange(-14,14,2),
               transform=ccrs.PlateCarree())
plt.colorbar(orientation='vertical', shrink=0.8, label='m/s')
plt.title('Vento Meridional (v)\nMédia 27 a 30 de Outubro de 2009')

plt.show()
```

```python
f=plt.figure(figsize=(10,6))

proj=ccrs.Orthographic(central_longitude=-60, central_latitude=-20, globe=None)
proj=ccrs.PlateCarree()
ax = plt.subplot(111, projection=ccrs.PlateCarree())
ax.coastlines(resolution='110m', linewidth=0.5)
plt.contourf(lon, lat, np.mean(myu, axis=0),
               extend='both',
               cmap = 'RdBu_r',
                levels=np.arange(-30,30,5),
               transform=ccrs.PlateCarree())
plt.colorbar(orientation='vertical', shrink=0.8, label='m/s')
plt.title('Vento Zonal (u)\nMédia 27 a 30 de Outubro de 2009')

plt.show()
```

### Climatologias (Médias de Longo Termo)

```python
u27 = np.mean(u[np.where(time.day==27)[0],:,:], axis=0)
u28 = np.mean(u[np.where(time.day==28)[0],:,:], axis=0)
u29 = np.mean(u[np.where(time.day==29)[0],:,:], axis=0)
u30 = np.mean(u[np.where(time.day==30)[0],:,:], axis=0)
```

```python
u27.shape
```

```python
uclim = (u27 + u28 + u29 + u30)/4
```

```python
f=plt.figure(figsize=(18,6))

proj=ccrs.Orthographic(central_longitude=-60, central_latitude=-20, globe=None)
proj=ccrs.PlateCarree()

ax = plt.subplot(131, projection=ccrs.PlateCarree())
ax.coastlines(resolution='110m', linewidth=0.5)
plt.contourf(lon, lat, uclim,
               extend='both',
               cmap = 'RdBu_r',
                levels=np.arange(-30,35,5),
               transform=ccrs.PlateCarree())
plt.colorbar(orientation='vertical', shrink=0.8, label='m/s')
plt.title('U 500hPa (1979 - 2019)\nMédia 27 a 30 de Outubro de 2009')

ax = plt.subplot(132, projection=ccrs.PlateCarree())
ax.coastlines(resolution='110m', linewidth=0.5)
plt.contourf(lon, lat, np.mean(myu, axis=0),
               extend='both',
               cmap = 'RdBu_r',
                levels=np.arange(-30,35,5),
               transform=ccrs.PlateCarree())
plt.colorbar(orientation='vertical', shrink=0.8, label='m/s')
plt.title('U 500hPa Episódio de ZCAS\nMédia 27 a 30 de Outubro de 2009')

ax = plt.subplot(133, projection=ccrs.PlateCarree())
ax.coastlines(resolution='110m', linewidth=0.5)
plt.contourf(lon, lat, np.mean(myu, axis=0) - uclim,
               extend='both',
               cmap = 'RdBu_r',
                levels=np.arange(-30,35,5),
               transform=ccrs.PlateCarree())
plt.colorbar(orientation='vertical', shrink=0.8, label='m/s')
plt.title('Diferença\nZCAS - Clima')

plt.show()
```
